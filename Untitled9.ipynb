```python
#4 . Dump all the data into car table 
with open("C:\\Users\\kapil\\Downloads\\car (1).data", "r") as myfile:
    text=myfile.read()

output_text = text.split(",")

with open("Output.txt", "w",) as outfile:
    for word in output_text:
        outfile.write(" '"+ word + "', " )
```

```python
import csv
with open("Output.txt" , 'r') as f:
    car_data=csv.reader(f,delimiter='\n')
    for i in car_data:
        print(type(i[0]))
        print(f'insert into car values ({str(i[0])})')
        cursor.execute('insert into cardataset.cars values({values})'.format(values = ((i[0]))))
```
